### 2. Cohésion des tableaux et convertion des valeurs

Ici nous voyons déjà que le code postal est du type float, qui ne comporte pas tous les éléments caractéristiques d'un code postal. Ces valeurs ne pourront pas être utilisées pour faire des études statistiques. Il nous faut donc les filtrer et convertir. Nous avons aussi les noms des variables qui ne sont pas identiques ("price", "prix") et les valeurs de certaines variables à modifier ("Appartement", "appartement"...). Tout ceci doit être traité en amont des études statistiques.

On va maintenant passer à la partie convertion.

```python
# on crée 2 fonctions pour convertir les objects en entier ou float en extrayant uniquement la valeur sans unité
# en entrée le data frame + les noms des variables à traiter, en sortie le data frame convertit
# 220 000.0 € str   ->  220000 int
# 45.5 m2     str   ->  45.5   float
# 75018.0     float -> '75018' string

def conversion_txt_int(data: pd.DataFrame, nom_var: Union[str, List[str]]) -> pd.DataFrame:
    # normaliser en liste pour toujours obtenir un DataFrame (et pas une Series)
    cols = [nom_var] if isinstance(nom_var, str) else list(nom_var)

    data[cols] = (
        data[cols]
            .apply(lambda s: (
                s.astype(str)
                 .str.extract(r"([\d\s]+)")[0]
                 .str.replace(" ", "", regex=False)
                 .replace("", pd.NA)
            ))
            .astype("Int64")
    )
    return data


def conversion_float_int_str(data: pd.DataFrame, nom_var: Union[str, List[str]]) -> pd.DataFrame:
    
    cols = [nom_var] if isinstance(nom_var, str) else list(nom_var)

    data[cols] = (
        data[cols]
          .apply(lambda s: (
              pd.to_numeric(s, errors="coerce")   # force float → numérique
                .astype("Int64")                  # enlève .0, garde NaN
                .astype("string")                 # conversion finale en string
          ))
    )
    return data



def conversion_txt_flt(data: pd.DataFrame, nom_var: Union[str, List[str]]) -> pd.DataFrame:

    # normaliser en liste
    if isinstance(nom_var, str):
        cols = [nom_var]
    else:
        cols = nom_var

    data[cols] = (
        data[cols]
            .apply(lambda s: (s.astype(str)
                 .str.extract(r"([\d\s]+(?:[.,]\d+)?)")[0]
                 .str.replace(" ", "", regex=False)
                 .str.replace(",", ".", regex=False)
                 .replace("", pd.NA)
            ))
            .astype(float)
    )

    return data
```

```python
# On fait des convertions et séparation unité et grandeur (120.5 m2 -> 120.5 ; 120 000.0 € -> 120000)
list_var1 = ["nombre_de_pieces", "prix", "Price_Per_M2"]
list_var2 = ["nombre_de_pieces", "prix"]
list_var3 = ["code_postal", "code_departement"]
list_var4 = ["surface_interieure"]

df_paris = conversion_txt_flt(df_paris, list_var4)
df_paris = conversion_float_int_str(df_paris, list_var3)
df_paris = conversion_txt_int(df_paris, list_var1)
df_etpro_2 = conversion_txt_int(df_etpro_2, list_var2)
df_etpro_1 = conversion_txt_int(df_etpro_1, list_var2)
```

### 3. Les filtres
Nous allons maintenant filtrer les incohérences présentes dans nos données.
Par exemple, nous souhaitons appliquer le prix au metre carre pour les données provenant de 'EtreProprio' (car non fourni).
Avant cela nous devons traiter les valeurs nulles pour la variable 'prix', les surfaces intérieures nulles ou encore les terrains avec une surface intérieure non nulle ...

```python
# Elimination des prix nulles, type de bien nulle, ville nulle, département nulle
df = df[df["prix"].notna()]
df = df[df["type_de_bien"].notna()]
df = df[df["code_departement"].notna()]
df = df[df["departement"].notna()]

print(f"Nombre d'annonce supprimées : {shape_df - df.shape[0]}")
shape_df = df.shape[0]
```

```python
df["type_de_bien"].unique()
```

```python
# faire une cohérence des valeurs
mapping = {
    "Appartement": "appartement",
    "Maison": "maison",
    "Terrain" : "terrain"
}
df["type_de_bien"] = df["type_de_bien"].replace(mapping)
```

```python
# Si pour les type de bien qui sont autre que 'terrain', la valeurs (prix) ou (surface intérieur) = 0, on supprimer la donnée

mask_invalide = (
    (df["type_de_bien"] != "terrain")
    & df["surface_interieure"].isna()
)

df = df[~mask_invalide]
# si le type de bien est un terrain et la surface inrtérieur n'est pas nulle, on supprime
df = df[~(
    (df["type_de_bien"] == "terrain")
    & df["surface_interieure"].notna()
)]

df = df[~(
    (df["type_de_bien"] == "terrain")
    & df["surface_terrain"].isna()
)]

print(f"Nombre d'annonce supprimées : {shape_df - df.shape[0]}")
shape_df = df.shape[0]
```

```python
# supprime les surfaces interieures < 10m2 et supérieur à 2000m2
types_avec_surface = ['maison', 'appartement',
                       'Duplex', 'Loft', 'Studio', 'Villa']
df = df[~(
    df["type_de_bien"].isin(types_avec_surface)
    & ((df["surface_interieure"] < 10)
    | (df["surface_interieure"] > 2000)
    )
)]

df = df[~(
    (df["type_de_bien"] == "terrain")
    & ((df["surface_exterieure"] < 50)
    | (df["surface_exterieure"] > 50000)
    )
)]


print(f"Nombre d'annonce supprimées : {shape_df - df.shape[0]}")
shape_df = df.shape[0]
```

```python
# on supprime les doublouns si présent
df = df.drop_duplicates(subset=["url_annonce"])
print(f"Nombre de doublons supprimés : {shape_df - df.shape[0]}")

```

```python
# on vérifie si le type de bien différent de 'terrain'

# On applique le calcule du prix surfacique

df.loc[df["type_de_bien"] == "terrain", "prix_m2"] = (
    df["prix"] / df["surface_terrain"]
).round(2)

df.loc[df["type_de_bien"] != "terrain", "prix_m2"] = (
    df["prix"] / df["surface_interieure"]
).round(2)
```

```python
# Pour finir, nous souhaitons filtrer les adresses afin quelle ne retourne uniquement la première partir 
# (sans le nom de la ville ni le code arrondissement)
df["Address"] = df["Address"].str.replace(
    r"\s*,?\s*Paris.*$", "", regex=True
)
```
