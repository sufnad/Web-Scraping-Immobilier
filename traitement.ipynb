```python
import pandas as pd         # permettant la création de fataframe
```

```python
# création csv
df_tl = pd.read_csv('annonces_tl.csv', sep = ';')
df_tc = pd.read_csv('annonces_tc.csv', sep = ';')
```

```python
df = pd.concat([df_tl, df_tc], ignore_index=True)
```

```python
nbr_bien = df.shape[0]
```

```python
df.info()
```

```python
# Sachons que normalement les variables tl et tc devrais chercher des biens de type terrain et commerce
# il n'est pas rare d'avoir d'autres types de biens dans ces recherches.
# on vérifie les incohérences le type de bien maison sans nombre de pièces

df = df[~(
    (df["type_de_bien"] == "maison") &
    (df["surface_interieure"].isna())
)]
df = df[~(
    (df["type_de_bien"] == "appartement") &
    (df["surface_interieure"].isna())
)]

df = df[~(
    (df["type_de_bien"] == "terrain") &
    (df["surface_interieure"].notna())
)]
```

```python
print(f'Nombre de biens exclu : {nbr_bien - df.shape[0]}')
```

```python
# on fait pareil avec les commerces
df = df[~(
    (df["type_de_bien"] == "maison") &
    (df["surface_interieure"].isna())
)]
df = df[~(
    (df["type_de_bien"] == "terrain") &
    (df["nombre_de_pieces"].notna())
)]

df = df[~(
    (df["type_de_bien"] == "terrain") &
    (df["surface_interieure"].notna())
)]

df = df[~(
    (df["type_de_bien"] == "terrain") &
    (df["surface_terrain"].isna())
)]

df = df[~(
    (df["type_de_bien"] == "commerce") &
    (df["surface_terrain"].notna())
)]

df = df[~(
    (df["type_de_bien"] == "appartement") &
    (df["surface_interieure"].isna())
)]

```

```python
print(f'Nombre de biens exclu : {nbr_bien - df.shape[0]}')
```
